```python
#Importa la librera OpenCV para el procesamiento de imgenes.
import cv2
#Importa la librera NumPy para operaciones numricas eficientes.
import numpy as np
#Importa cv2_imshow para mostrar imgenes en Google Colab.
from google.colab.patches import cv2_imshow

#Define una funcin llamada getContours para detectar y analizar contornos en una imagen.
def getContours(img, imgCountours):
  #Encuentra todos los contornos externos en la imagen y los almacena en 'contours' y su jerarqua en 'hierarchy'.
  contours, hierarchy = cv2.findContours(img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
  #Itera sobre cada contorno encontrado.
  for cnt in contours:
    #Calcula el rea del contorno actual.
    area = cv2.contourArea(cnt)
    #Imprime el rea del contorno.
    print(area)
    #Si el rea del contorno es mayor que 500 pxeles:
    if area > 600:
      #Dibuja el contorno en la imagen 'imgCountours' en rojo.
      cv2.drawContours(imgCountours, cnt, -1, (0, 0, 255), 2)
      #Calcula el permetro del contorno.
      #Imprime el permetro del contorno.
      print(cv2.arcLength(cnt, True))

      #Aproxima el contorno con una precisin del 1.76% de su permetro.
      approx = cv2.approxPolyDP(cnt, 0.0176*cv2.arcLength(cnt, True), True)
      #Imprime el nmero de vrtices del contorno aproximado.
      print(len(approx))
      #Obtiene el nmero de vrtices del contorno aproximado.
      objCor = len(approx)
      #Obtiene las coordenadas y dimensiones del rectngulo delimitador del contorno.
      x, y, w, h = cv2.boundingRect(approx)

      # Determina el tipo de figura geomtrica segn el nmero de vrtices.
      objectType = ""
      #Si tiene 3 vrtices...
      if objCor == 3:
        objectType = "Triangle"
      #Si tiene 4 vrtices...
      elif objCor == 4:
        #Calcula la relacin de aspecto del rectngulo delimitador.
        aspRatio = w/float(h)
        #Si la relacin de aspecto est cerca de 1, es un cuadrado.
        if aspRatio > 0.95 and aspRatio < 1.05:
          objectType = "Square"
        #De lo contrario, es un rectngulo.
        else:
          objectType = "Rectangle"
      #Si tiene 5 vrtices...
      elif objCor == 5:
        objectType = "Pentagon"
      #Si tiene 6 vrtices...
      elif objCor == 6:
        objectType = "Hexagon"
      #Si tiene 7 vrtices...
      elif objCor == 7:
        objectType = "Heptagon"
      #Si tiene 8 vrtices...
      else:
         #Calcula la relacin de aspecto del rectngulo delimitador.
         aspRatio = w/float(h)
         #Si la relacin de aspecto est cerca de 1, es un crculo.
         if aspRatio > 0.95 and aspRatio < 1.05:
          objectType = "Circle"
         #De lo contrario, es una elipse.
         else:
          objectType = "Ellipse"

      #Escribe el nombre de la figura al lado del contorno.
      if objectType:
        cv2.putText(imgCountours, objectType, (x+(w//2)-10, y+(h//2)-10), cv2.FONT_HERSHEY_PLAIN, 1, (0,0,0),2)

#Se lee la imagen (se tiene que especificar la ruta).
img = cv2.imread("/content/fig.png")
img = cv2.resize(img, (0,0), None, 0.8, 0.8)
if img is None:
    print("Error: No se pudo cargar la imagen.")
else:
    imgCountours = img.copy()
    #Se pone la imagen en escala de grises.
    imgGray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    #Se pone la imagen borrosa.
    imgBlur = cv2.GaussianBlur(imgGray, (7, 7), 1)
    #Se pone la imagen con contraste (negro y blanco).
    imgCanny = cv2.Canny(img, 100, 100)
    #Se le aade el contorno.
    getContours(imgCanny, imgCountours)


#Se muestra la imagen original.
cv2_imshow(img)
#Se muestra la imagen en escala de grises.
cv2_imshow(imgGray)
#Se muestra la imagen borrosa.
cv2_imshow(imgBlur)
#Se muestra la imagen con contraste.
cv2_imshow(imgCanny)
#Muestra la imagen con el contorno.
cv2_imshow(imgCountours)
```
