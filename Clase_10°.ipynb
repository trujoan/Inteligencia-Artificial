Clustering

```python
# Se crea en este caso una caja de zapatos que va a tener varios elementos dentro de él

Caja_de_zapatos = [
    {"color": "Rosa", "talla": "grande"},
    {"color": "Morado", "talla": "mediano"},
    {"color": "Negro", "talla": "grande"},
    {"color": "Rosa", "talla": "pequeño"}
]

# Aquí se segmentan por color esos elementos en este caso es es un cluster que es cómo las características que mirabamos en los elementos el cluster es de acuerdo al color
grupos_por_color = {}

for zapato in Caja_de_zapatos:
    color = zapato["color"]
    if color not in grupos_por_color:
        grupos_por_color[color] = []
    grupos_por_color[color].append(zapato)

# Se imprimen los grupos
for color, zapatos in grupos_por_color.items():
    print("Grupo de zapatos", color, ":")
    for zapato in zapatos:
        print("-", zapato)
```

GMM

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture

# Generate data
np.random.seed(42)
X = np.random.multivariate_normal([0, 0], [[1, 2], [2, 1]], size=100)

# Fit GMM model
gmm = GaussianMixture(n_components=2)
gmm.fit(X)

# Predict cluster labels
labels = gmm.predict(X)

# Plot data and clusters
plt.scatter(X[:, 0], X[:, 1], c=labels, s=5)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('GMM clustering')
plt.show()
```

kmeans

```python
# Importar la biblioteca necesaria
from sklearn.cluster import KMeans

# Datos de ejemplo
X = [[1], [2], [3], [6], [7], [8]]

# Crear el objeto KMeans
kmeans = KMeans(n_clusters=2, random_state=0)

# Ajustar el modelo a los datos
kmeans.fit(X)

# Obtener las etiquetas de cluster asignadas
labels = kmeans.labels_

# Obtener los centroides
centroids = kmeans.cluster_centers_

# Imprimir las etiquetas y los centroides
print("Etiquetas de cluster:", labels)
print("Centroides:", centroids)
```

Optical

```python
!pip install opencv-python
import cv2
import numpy as np

# Lee dos imágenes
Imagen1 = cv2.imread('Programacion/Imagen1.jpg')
Imagen2 = cv2.imread('Programacion/Imagen2.jpg')

# Las convierte en escala de grises
gray1 = cv2.cvtColor(Imagen1, cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(Imagen2, cv2.COLOR_BGR2GRAY)

# Calcula el flujo óptico entre ambas
flow = cv2.calcOpticalFlowFarneback(gray1, gray2, None, 0.5, 3, 15, 3, 5, 1.2, 0)

# Crea una rueda de colores para visualizar la dirección
color_wheel = np.zeros((512, 512, 3), dtype=np.uint8)
for i in range(512):
    color_wheel[i, :, 0] = 255
    color_wheel[i, :, 1] = np.uint8(255 * (i / 256.0))
    color_wheel[i, :, 2] = np.uint8(255 * (1.0 - i / 256.0))

# Convierte el flujo en coordinadas polares que tienen magnitud y ángulo
magnitude, angle = cv2.cartToPolar(flow[..., 0], flow[..., 1])

# Une el ángulo con la rueda de colores (para saber el color)
color = color_wheel[np.int32(angle * (512 / (2 * np.pi)))]

# Normaliza la magnitud para que estén dentro del rango
magnitude = np.uint8(np.clip(magnitude, 0, 255))

# Combina el color y las magnitudes en una sola imagen
flow_image = cv2.merge([color, color, magnitude])

# Convierte la imagen en RGB
flow_image_bgr = cv2.cvtColor(flow_image, cv2.COLOR_HSV2BGR)

# Muestra la imagen del flujo
cv2.imshow('Flow', flow_image_bgr)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
